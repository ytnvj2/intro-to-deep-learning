```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
N=500
D=2
np.random.seed(100)
X1=np.random.randn(N,D)+np.random.randn()*3
X2=np.random.randn(N,D)-np.random.randn()*4
X=np.concatenate((X1,X2))
y=np.array([0]*N+[1]*N)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
W=tf.Variable(tf.random_normal((2,),stddev=np.sqrt(D)))
b=tf.Variable(tf.random_normal((1,)))
```

```python
tfX=tf.placeholder(np.float32,[None,D])
tfY=tf.placeholder(np.float32,[None,])
y_p=tf.sigmoid(tf.tensordot(tfX,W,axes=1)+b)
```

```python
cost=tf.reduce_sum(tf.nn.sigmoid_cross_entropy_with_logits(labels=tfY,logits=y_p))
```

```python
train=tf.train.GradientDescentOptimizer(0.001).minimize(cost)
preds=tf.round(y_p)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    c=[]
    for i in range(1000):
#         s.run(y_p,feed_dict={tfX:X})
        c.append(sess.run(cost,feed_dict={tfX:X,tfY:y}))
        sess.run(train,feed_dict={tfX:X,tfY:y})
        p=sess.run(preds,feed_dict={tfX:X})
        print(np.mean(y==p))
```
