```python
import numpy as np
import pandas as pd
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
df=pd.read_csv('data_2d.csv',sep=';')
```

```python
df['X0']=1
df=df[['X0','Speed','Angle','Power']]
```

```python
df['Power'].replace(',','.',method='')
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
a=[]
for i in y:
    a.append(i.replace(',','.'))
y=np.array(a)
y=y.astype(np.float64)
```

```python
y.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,1],X[:,2],y)
plt.show()
```

```python
#Solving the Linear coefficients
B = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))
```

```python
np.dot(X.T,y).shape
```

```python
y_pred=X.dot(B)
```

```python
y_pred
```

```python
RSS=(y-y_pred)**2
TSS=(y-y.mean())**2
R2=1-(RSS.sum()/TSS.sum())
```
