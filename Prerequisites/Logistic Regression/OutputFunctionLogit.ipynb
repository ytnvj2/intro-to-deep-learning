```python
import numpy as np
import matplotlib.pyplot as plt
N=100
D=2
X=np.random.randn(N,D)
ones=np.array([np.ones(N)]).T
Xb=np.concatenate((ones,X),axis=1)
```

```python
w=np.random.randn(D+1)/np.sqrt(2)
```

```python
z=Xb.dot(w)
```
