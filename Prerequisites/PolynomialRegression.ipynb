```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(1,100,50)
y=X**4+X.sum()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(X)
```

```python
df['X2']=X**2
df['X0']=1
```

```python
df=df[['X0',0,'X2']]
```

```python
X=df.values
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,1],X[:,2],y)
plt.show()
```

```python
# weight matrix
B=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))
```

```python
y_pred=X.dot(B)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,1],X[:,2],y)
ax.plot(X[:,1],X[:,2],y_pred)
plt.show()
```

```python
RSS=(y-y_pred)**2
TSS=(y-y.mean())**2
R2=1-(RSS.sum()/TSS.sum())
```
