```python
import numpy as np
```

```python
X=np.linspace(1,100,50)
y=X**1.25+X.sum()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Calculating the coefficients B0 and B1
# such that y = B0 + B1*X
B0=(y.mean()*(X.dot(X)/len(X))-(X.mean()*(X.dot(y)/len(X))))/((X.dot(X)/len(X))-(X.mean()**2))
B1=(X.dot(y)/len(X)-(X.mean()*y.mean()))/((X.dot(X)/len(X))-(X.mean()**2))
```

```python
B0
```

```python
B1
```

```python
X.sum()
```

```python
y_pred=B0+B1*X
```

```python
plt.plot(X,y_pred)
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y pred')
plt.show()
```
