```python
import numpy as np
import matplotlib.pyplot as plt
N=50
D=50
X=(np.random.random((N,D))-0.5)*10
```

```python
X.shape
```

```python
true_B=np.array([1,0.5,-0.5,0.75]+[0]*(D-4))
y=X.dot(true_B)+np.random.randn(N)*0.5
```

```python
B=np.random.randn(D)/np.sqrt(D)
lr=0.001
l1=10
cost=[]
for i in range(1000):
    B=B-lr*(X.T.dot(X.dot(B)-y) + l1*np.sign(B))
    y_pred=X.dot(B)
    RSS=(y-y_pred)**2
    cost.append(RSS.sum()/len(y))
```

```python
B
```

```python
plt.plot(y)
plt.plot(y_pred)
plt.show()
```

```python
TSS=(y-y.mean())**2
1-(RSS.sum()/TSS.sum())
```
