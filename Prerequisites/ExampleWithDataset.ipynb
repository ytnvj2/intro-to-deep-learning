```python
# Calculate Weights
B=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))
```

```python
B
```

```python
y_pred=X.dot(B)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],y)
ax.plot(X[:,0],X[:,1],y_pred)
plt.show()
```

```python
def get_R2(X,y):
    B=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))
    y_pred=X.dot(B)
    RSS=(y-y_pred)**2
    TSS=(y-y.mean())**2
    R2=1-(RSS.sum()/TSS.sum())
    print('R-squared is',R2)
```

```python
get_R2(X[:,:],y)
```

```python
get_R2(X[:,0:2],y)
```

```python
get_R2(X[:,1:],y)
```

```python
get_R2(X[:,0:1],y)
```
