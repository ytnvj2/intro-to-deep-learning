```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(0,10,50)
y=0.5*X + np.random.randn(50)
y[-1]+=30
y[-3]+=40
```

```python
X
```

```python
y
```

```python
plt.scatter(X,y)
plt.show()
```

```python
X=np.vstack([np.ones(len(X)),X]).T
```

```python
B_ml=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))
y_pred_ml=X.dot(B_ml)
plt.scatter(X[:,1],y)
plt.plot(X[:,1],y_pred_ml)
plt.show()
```

```python
l2=1000.00
B_map=np.linalg.solve(l2*np.eye(2)+np.dot(X.T,X),np.dot(X.T,y))
y_pred_map=X.dot(B_map)
plt.scatter(X[:,1],y)
plt.plot(X[:,1],y_pred_ml)
plt.plot(X[:,1],y_pred_map)
plt.show()
```
